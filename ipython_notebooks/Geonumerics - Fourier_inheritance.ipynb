```python
import pylab as py
import numpy as np
def FourierDerivative(f):
    N = np.size(f)
    n = np.arange(0,N)
    # df discrete differential operator
    df = np.complex(0,1)*py.fftshift(n-N/2)
    dfdt = py.ifft( df*py.fft(f) )  
    return py.real(dfdt)
```

```python
#wd2d.Source(0)

# calculate derivatives using fourier
# derivatives in x and y direction
d2dx = np.zeros([wd2d.Nz, wd2d.Nx]) 
d2dz = np.zeros([wd2d.Nz, wd2d.Nx])

# xline by xline
for i in range(wd2d.Nz):        
    d2dx[i] = FourierDerivative(wd2d.Ucurrent[i]) # first derivative
    d2dx[i] = FourierDerivative(d2dx[i]/wd2d.Rho[i]) # second with Rho

# to easy calculate on the z direction, transpose the matrixes
wd2d.Ucurrent = wd2d.Ucurrent.transpose()
wd2d.Rho = wd2d.Rho.transpose()
d2dz = d2dz.transpose()
# zline by zline
for i in range(wd2d.Nx):        
    d2dz[i] = FourierDerivative(wd2d.Ucurrent[i]) # first derivative
    d2dz[i] = FourierDerivative(d2dz[i]/wd2d.Rho[i]) # second with Rho
# transpose back
wd2d.Ucurrent = wd2d.Ucurrent.transpose()
wd2d.Rho = wd2d.Rho.transpose()
d2dz = d2dz.transpose()

# derivatives combination + source
LP = d2dx + d2dz      
# simple centered differences in time 2nd order
# matrix operations
wd2d.Ufuture = ((wd2d.Vel*wd2d.Dt)**2)*LP*wd2d.Rho+2*wd2d.Ucurrent-wd2d.Uprevious     

# update stack times
wd2d.Uprevious = wd2d.Ucurrent
wd2d.Ucurrent = wd2d.Ufuture
imshow(wd2d.Ucurrent)
```
