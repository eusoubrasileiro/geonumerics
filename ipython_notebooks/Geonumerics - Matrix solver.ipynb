```python
import scipy.linalg as ln
def lulu(wv2d, mUt):
    mUtfactor = ln.lu_factor(mUt)
    print shape(mUtfactor)
    wv2d.iter=1
    wv2d.Source(wv2d.iter)
    v = wv2d.Independent()
    result = ln.lu_solve(mUtfactor, v)
    return result
    
def chocho(wv2d, mUt):
    mUtfactor = ln.cho_factor(mUt, overwrite_a=True)
    print shape(mUtfactor)
    wv2d.iter=1
    wv2d.Source(wv2d.iter)
    v = wv2d.Independent()
    result = ln.cho_solve(mUtfactor, v)
    return result
```
