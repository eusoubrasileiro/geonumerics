```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
from scipy.integrate import cumtrapz

#import seaborn as sns

from __future__ import division

from scipy.signal import bartlett
```

```python
f1 = 4.
f2 = 86.
SL = 25.
db_oct = -1.

n = (db_oct)/6. +1
k1 = np.power(f1,n)
k2 = (np.power(f2,n) - np.power(f1,n))/SL

def sr(x):
    return np.power((k1 + k2*x),1./n)

def linear(x):
    return f1 + (f2 - f1)/SL*x
```

```python
t = np.linspace(0,25,101)

plt.figure()

plt.plot(t,sr(t), label='dB/oct')
plt.plot(t,linear(t), label='linear')
plt.legend(loc='upper left')
plt.xlabel('tempo de sweep (s)')
plt.ylabel(u'frequência (Hz)')
plt.show()
plt.close()
```

```python
def sweep(x):
    f = cumtrapz(sr(x),x, initial=f1)
    phase = 2*np.pi*cumtrapz(f,x, initial=0)
    sweep = np.cos(phase)
    return sweep

def sweep_teorico(x):
    def freq(T):
        return (np.power((k2*T+k1),(1/n))*(k2*n*T+k1*n))/(k2*n+k2)-(np.power(k1,(1/n+1))*n)/(k2*n+k2)
    
    phase = 2*np.pi*cumtrapz(freq(x),x, initial=0)
    sweep = np.cos(phase)
    return sweep
```

```python
t = np.linspace(0,25,12500)

plt.figure(figsize=(15,3))

plt.plot(t,sweep_teorico(t), label='analitico')
plt.plot(t,sweep(t), label='numerico')

plt.legend(loc='upper left')
plt.xlabel('tempo de sweep (s)')
plt.ylabel(u'Amplitude')

plt.show()
plt.close()
```

```python
plt.figure(figsize=(15,3))

bar1 = bartlett(1/0.002)[:250]
bar2 = bartlett(1/0.002)[250:]
one = np.ones(12000)
tapper = np.concatenate((bar1,one,bar2))

plt.plot(t,tapper, label='trapezoidal')

plt.legend(loc='upper left')
plt.xlabel('tempo de sweep (s)')
plt.ylabel(u'Amplitude')

plt.show()
plt.close()
```

```python
t = np.linspace(0,25,12500)

plt.figure(figsize=(15,3))

plt.plot(t,sweep_teorico(t)*tapper, label='analitico')

plt.xlabel('tempo de sweep (s)')
plt.ylabel(u'Amplitude')

plt.show()
plt.close()
```
