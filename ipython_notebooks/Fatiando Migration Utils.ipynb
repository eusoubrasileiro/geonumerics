```python
from PIL import Image
import numpy as np
img = Image.open('wedge.tif')
img.load()
img = img.convert('I') # gray scale, convert format
data  = np.asarray(img, dtype=np.float32)
data[:][data[:] == 255.] = 2500.0 # overburden
data[:][data[:] == 0. ] = 3500 # the triangle
data[:][data[:] == 146.] = 2300.0 # the first slab bellow the triangle
data[:][data[:] == 36.] = 2700.0 # the last slab
cshape = data.shape
print cshape
```

```python
%pylab inline
imshow(data)
```

```python
%cd fatiando
from fatiando.seismic import wavefd
# Set the parameters of the finite difference grid
dx = 50. # spacing
dz = 25.
area = [0, cshape[1]*dx, 0, cshape[0]*dz]
# Set the parameters of the finite difference grid
velocity = data
# avoiding spatial alias, frequency of source should be smaller than this
ds = dx
if dz > ds:
    ds = dz
fc = 0.5*np.min(velocity)/dz  # based on plane waves v=l*f
fc -= 0.5*fc # a little smaller
dt = wavefd.scalar_maxdt(area, cshape, np.max(velocity))
duration = 4.0
maxit = int(duration/dt)
#stations = [[125*dx, 0*dz]] # x, z coordinate of the seismometer TODO: take a look why need to be buried so deep... 
# to record
stations = [[ 50*dx + 2*dx*i, 50.0] for i in xrange(100)]
sources = [wavefd.GaussSource((150*dx, 5*dz), area, cshape,  1., fc)]
snapshots = 3 # every 3 iterations plots one
simulation = wavefd.scalar(velocity, area, dt, maxit, sources, stations, snapshots)
```
