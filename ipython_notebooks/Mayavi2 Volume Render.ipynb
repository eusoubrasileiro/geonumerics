##Plotting simple simulation time step

```python
from mayavi import mlab

fig = mlab.figure(size=(600,600))
sscalar = mlab.pipeline.scalar_field(acustic3_esg)
svolume = mlab.pipeline.volume(sscalar)
svolume.volume_mapper.lock_sample_distance_to_input_spacing = True
svolume.volume_mapper.blend_mode = 'maximum_intensity'
# maps a value in a rgba value (colormap)
# rgb to a value map is made trough the ColorTransferFunction - ctf
# value to opacity is made trough OpacityTransferFunction - otf 
# the values on simulation vary from [-something, 0, +something] 
# is a zero centric palete, so zero should be opaque
minu = acustic3_esg.min(); maxu = acustic3_esg.max()
# changing colorbar! first (RedWhiteBlue)
from tvtk.util.ctf import ColorTransferFunction
ctf = ColorTransferFunction()
#ctf.add_rgb_point(value, r, g, b)  # r, g, and b are float between 0 and 1
ctf.add_rgb_point(minu, 1., 0., 0.)
ctf.add_rgb_point(0., 0., 0., 0.) 
ctf.add_rgb_point(maxu, 0., 0., 1.) 
svolume._volume_property.set_color(ctf)
svolume._ctf = ctf
svolume.update_ctf = True
# Changing the otf: (opacity correspondence) also zero centric
from tvtk.util.ctf import PiecewiseFunction
otf = PiecewiseFunction()
otf.add_point(minu, 0.4)
otf.add_point(0., 0.0)
otf.add_point(maxu, 0.4)
svolume._otf = otf
svolume._volume_property.set_scalar_opacity(otf)
mlab.show()  # make possible to manipulate at the end
```

## Trying to create a custom colormap or in mayavy vtk called look up table (LUT)
More references in [here](http://docs.enthought.com/mayavi/mayavi/auto/example_custom_colormap.html#example-custom-colormap) for adding the look up table

And [here](http://matplotlib.org/api/colors_api.html#matplotlib.colors.LinearSegmentedColormap) for trying to copy a RdBu from matplotlib

```python
from mayavi import mlab

fig = mlab.figure(size=(600,600))
sscalar = mlab.pipeline.scalar_field(acustic3_esg)
svolume = mlab.pipeline.volume(sscalar)
svolume.volume_mapper.lock_sample_distance_to_input_spacing = True
svolume.volume_mapper.blend_mode = 'maximum_intensity'
# maps a value in a rgba value (colormap)
# rgb to a value map is made trough the ColorTransferFunction - ctf
# value to opacity is made trough OpacityTransferFunction - otf 
# the values on simulation vary from [-something, 0, +something] 
# is a zero centric palete, so zero should be opaque
minu = acustic3_esg.min(); maxu = acustic3_esg.max()
# changing colorbar! first (RedWhiteBlue)
lut = svolume.module_manager.scalar_lut_manager.lut.table.to_array()

# Changing the otf: (opacity correspondence) also zero centric
from tvtk.util.ctf import PiecewiseFunction
otf = PiecewiseFunction()
otf.add_point(minu, 0.4)
otf.add_point(0., 0.0)
otf.add_point(maxu, 0.4)
svolume._otf = otf
svolume._volume_property.set_scalar_opacity(otf)
mlab.show()  # make possible to manipulate at the end
```
