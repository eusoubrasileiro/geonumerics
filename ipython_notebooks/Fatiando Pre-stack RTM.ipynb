###Make first arrival time index map###

This is the same approach the McMechan did, altough he doesn't explicitly say how he did for finding the first arrivals time. 

Could be eikonal equation for making this map but, this would be different from McMechan approachs. 
Or for **correlation image condition** it is not necessary.

Have to think and try a better methodology for doing this first arrival analysis

For while using the cross-correlation with itself inverted and getting the peak of maximum value.

This is misleading for places where the wave trajectory made the first arrival be smaller than the subsequent.

```python
from scipy.signal import fftconvolve
import numpy as np

def fftcorrelate(b, a):
    if b.size%2 != 0: # acceptable error
        b_ = np.append(b, np.zeros(1))
    b_ = numpy.zeros(b.size * 2)
    b_[b.size/2:b.size/2+b.size] = b # This works for data_length being even
    # Do an array flipped convolution, which is a correlation.
    c = fftconvolve(b_, a[::-1]) 
    return c[::-1][1+b.size/2:]

def getarrival_index(trc):
    r"""
    get the first arrival time index
    cross-correlate the trace with itself inverted 
    and get the maximum correlation index    
    """
    autotrc = fftcorrelate(trc[::-1], trc)
    return (np.argmax(autotrc))/2    

trc = stspsfwd[:, 65, 5]
figure(figsize=(5,5))
plot(trc, 'b')
plot(trc[::-1],'r')
imax = getarrival_index(trc)
plot(getarrival_index(trc), 0.0, 'or')
print imax
```

```python
time_map = np.zeros((160,70))

for k in xrange(160):
    for i in xrange(70):
        time_map[k, i] = getarrival_index(stspsfwd[:, k, i])
```

Clean the unusable traces to be ready for stacking!

```python
def zmutelinear(data, t0i, dkdi):
    ntrc = data.shape[1]    
    zmax = data.shape[0]
    for i in xrange(ntrc):
        imute = t0i + int(round(dkdi*i))
        if imute < 0 or imute > zmax:
            break
        data[0:imute, i] = 0.0
    return data
```

```python
umigrated[:,:35] = zmutelinear(umigrated[:,:35], 140, -14./3.)
umigrated[:,35:] = zmutelinear(umigrated[:,35:], 0, 14./3.)
figure(figsize=(15,10))
imshow(umigrated, aspect=0.32, origin='upper', cmap=mpl.cm.gray_r, vmin=-.001, vmax=.001)
```
