##Apply NMO correction on synthetic data shots ##

###First on a cmp gather then on the entire set ###

For a given layered earth with $N$ layers with interval velocities given by:

$$ v_i = [ v_0, v_1, v_2, v_3 ... v_n] $$

And likewise with thickness given by:

$$ d_i = [ d_0, d_1, d_2, d_3 ... d_n] $$

The root mean square velocity at the $i$th layer down in the earth is given by:

\begin{equation*} V^2_{{rms}_i} = \frac{1}{\left(\sum_{j=0}^i \Delta t_j \right)}   \sum_{j=0}^i v_j^2 \Delta t_j \end{equation*}

Where the two way time $\Delta t_j$ :

$$ \Delta t_j = 2\frac{v_j}{d_j} $$

*For a horizontal stratified earth the $V_{rms}$ is equal the $V_{nmo}$ for any layer for a small spread aproximation
(offset small compared to depth).*

So the hyperbolic normal moveout correction for a  $i$th layer is given by:

$$ t_i^2 = t_{0_i}^2 + \frac{x^2}{V_{{rms}_i}} $$

```python
def vrms(vi, ds):
    """
    Calculate RMS velocity from:

    * vi : ndarray
        interval velocity
    * ds : ndarray
        layer size
    
    return Rms velocity array
    """
    twt = 2*(ds/vi) # two way time
    vi2_twt = (vi**2)*twt
    return np.sqrt(vi2_twt.cumsum()/twt.cumsum())    
```

```python
dx = 50.0 # from grid size
xshots = [] 
for i in xrange(100):
    sx = 5000.0 + 2*dx*i
    st0 = sx - 3550 # first station (Stack Array)
    stations = [ st0 + 2*dx*i for i in xrange(70)]
    xshots.append([sx, stations]) 

nsamples = 917
dummy = np.zeros(70)

figure(figsize=(10,10))

for i in xrange(100):
    plot(xshots[i][1], dummy+i, 'g+')  # stations
    plot(xshots[i][0], i, 'r*')  # source

ylabel("shot number")
xlabel("line position (m)")
pylab.show()
```

###Sort all traces by cmp###

Working just with trace header information : geometry information.

Create cmp value of each trace

```python
cmps = [] 
offsets = [] 
for i in xrange(100):
    cmps.append([ (xshots[i][1][j]+xshots[i][0])/2. for j in xrange(70)])    
    offsets.append([ abs(xshots[i][1][j]-xshots[i][0]) for j in xrange(70)])   

# create a list of all traces including all header information needed: cmp, offset, xcoord, trace number
htraces = []
for i in xrange(100):
    for j in xrange(70):
        htraces.append((cmps[i][j], offsets[i][j], xshots[i][1], i*70+j)) 
        
# sort by cmp
htraces.sort(key=lambda header: header[0])

import pandas

figure(figsize=(10,5))
scmp = pandas.Series([htraces[i][0] for i in xrange(7000)])
print scmp.nunique() # number o unique cmps
scmp.hist(bins=scmp.nunique())
ylabel('cmp fold')
xlabel('cmp line position (m)')
```

###Apply NMO Correction in all cmp traces###

```python
from scipy.interpolate import UnivariateSpline as ispline

def index_vmodel(xcoord): # from dx = 50 simulation grid
    return xcoord/50

cmp_xcoords = scmp.unique()
ds = np.ones(160)*25 # from dz = 25 simulation grid
dt=0.0043575828308287

cmps_stacked = np.zeros((917, ncmps))

for i in xrange(ncmps):
    vi = vdata[:, index_vmodel(cmp_xcoords[i])]    
    vrmsi = vrms(vi, ds)    
    # linear with extrapolation possible, cubic is bad at end and start
    fs = ispline(twt, vrmsi, bbox=[0.0, 917*dt], k=1) 
    ntwt = np.linspace(0.0, 917*dt, 917)
    svrms = fs(ntwt)    
    cmps[:180, :, i] = 0.0 # mute all instead of fk for direct arrivals
    svrms *= 0.92 # small tuning for better stacks
    cmp_nmo = utils.nmo(svrms, cmps[:, :, i], offsets[:, i], dt)
    #figure(figsize=(10,7))
    #pylab.imshow(cmp_nmo, aspect=0.06, origin='upper', cmap=mpl.cm.gray_r)
    cmps_stacked[:, i] = cmp_nmo.sum(1) # stack all traces
```
