**Testing running without absourbing boundary conditions**

I had to modify the code to not call `_nonreflexive_scalar_boundary_conditions` and then run again.

```python
import sys
import numpy as np
from fatiando.seismic import wavefd

# Set the parameters of the finite difference grid 2D
shape = (50, 50)
ds = 15.  # spacing
area = [0, shape[0]*ds, 0, shape[1]*ds]
# Set the parameters of the finite difference grid
velocity = np.ones(shape)*1500.  # m/s
# avoiding spatial alias, frequency of source should be smaller than this
fc = 0.5*np.min(velocity)/ds  # based on plane waves v=l*f
fc -= 0.5*fc
sources = [wavefd.GaussSource((25*ds, 0*ds), area, shape,  1., fc)]
dt = wavefd.scalar_maxdt(area, shape, np.max(velocity))
duration = 0.8
maxit = int(duration/dt)
# x, y, z coordinate of the seismometer
#stations = [[45*ds, 45*ds, 65*ds], [65*ds, 65*ds, 30*ds]]
stations = []
snapshots = 15  # every 1 iterations plots one
simulation = wavefd.scalar(velocity, area, dt, maxit, sources, stations, snapshots)
movie = np.zeros(((maxit/snapshots)+2, 50, 50))
i = 0
for t, u, seismogram in simulation:
    movie[i] = u
    sys.stdout.write("\rprogressing .. %.1f%% time %.3f"%(100.0*float(t)/maxit, (dt*t)))
    sys.stdout.flush()
    i += 1
```

** Using pylab to make QC plots **

Necessary note:

* There is a padding added to the grid that is by default 5% of nz, there is where the wave vanishes

```python
%pylab inline
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 10, 9 # that's default image size for this interactive session
print np.shape(movie)
#iter = 7
minv = -10**(-1)
maxv = 10**(-1)
for i in range(movie.shape[0]):
    figure()
    imshow(movie[i], vmin=minv, vmax=maxv, cmap=mpl.cm.gray_r, origin='upper')

without_abs = movie
```
